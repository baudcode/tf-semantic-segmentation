```python
import math
import time
from tf_semantic_segmentation.bin.train import get_args, train_test_model
from pprint import pprint

from tf_semantic_segmentation.models.deeplabv3 import deeplabv3
from tf_semantic_segmentation.models import models_by_name
from tf_semantic_segmentation.models.deeplabv3plus import deeplabv3_plus_xception, deeplabv3_plus_mobile
from tensorflow.keras import backend as K

# shutil.rmtree('logs-all')

args = get_args({})

args.data_dir = './data'
args.model = "unet"
args.augmentations = []

args.buffer_size = 100
args.val_buffer_size = 20

args.dataset = 'shapes'
args.train_on_generator = True
args.size = [256, 256]
args.resize_method = 'resize' # use normal resize
args.validation_freq = 100 # validate every 10 epochs

args.mixed_float16 = False
args.no_tensorboard = True # do not log other images/metrics
args.no_evaluate = True
args.no_early_stopping = True

# learning rate and optimizer
args.learning_rate = 2e-4
args.optimizer = 'adam' # ['adam', 'radam', 'ranger']
# args.tensorboard_val_images = True


# loss and corresponding activation function
args.loss = 'categorical_crossentropy'
args.final_activation = "softmax"
args.metrics = ['f1_score', 'iou_score']

# args.model_weights = 'logs/camvid-erfnet-bs4-e300-lr0.0001-adam-categorical_crossentropy-softmax/best-weights.h5'
args.tensorboard_update_freq = 'batch'


args.logdir = logdir
# args.record_dir = "records/%s" % dataset

args.steps_per_epoch = 32
args.batch_size = 4
args.epochs = 15
args.find_lr = True # logs metrics to tensorboard
args.find_lr_min_lr = 1e-9
args.find_lr_max_lr = 1e-1
args.find_lr_beta = 0.98 # smoothing factor
args.find_lr_stop_factor = 4 # stop training when loss increases 4x best loss

results, model = train_test_model(args)
```
